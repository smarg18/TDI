## Plot 1

Avalanches kill dozens of people in the United States each year, and these numbers have increased in the past two decades with the increasing popularity and accessibility of backcountry skiing and touring. Despite significant efforts to track and understand avalanches, these natural events remain difficult to predict even in known high-risk terrain due to complex contributing factors including vegetation, temperature, preciptation, snow depth, wind, and other factors.

This complexity, coupled with the availability of vast amounts of climate data from the US Government, presents the opportunity to use machine learning to parse the trends.

First, a look at the numbers motivating this research.

```python
import numpy as np
import pandas as pd
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as sm
```

```python
d = pd.read_excel('CAICavalanche.xls')
```

```python
from pandas import Series, DataFrame
deaths = d.groupby(['Avalanche Year'])['KL'].sum()
```

```python
sns.set()
x = d['Avalanche Year']
num_bins = len(deaths)
plt.hist(x, num_bins, normed=1, facecolor=sns.xkcd_rgb["denim blue"], alpha=0.8)
plt.xlim([1951,2017])
plt.xlabel('Avalanche Year')
plt.ylabel('Fatalities')
plt.title(r'Fatalities per Avalanche Year: $1951 - 2017$')
plt.subplots_adjust(left=0.2)
plt.show()
```

```python
sns.set()
k = d.groupby(['Travel Mode']).size().reset_index(name = 'Counts')
labels = k['Travel Mode']
sizes = k['Counts']

plt.pie(sizes, labels=labels, shadow=True, startangle=140)
 
plt.axis('equal')
plt.show()
```
