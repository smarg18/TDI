Existing research and logic both suggest that avalanches are caused in part by weather conditions such as snowfall; however, this relationship is particularly complex and may be mitigated or exacerbated by many other factors.

Below, as motivation for additional research, this chart suggests that avalanches can occur with significantly higher snowfall than normal or higher accumulated snow depth than normal.

```python
import numpy as np
import pandas as pd
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import Series, DataFrame
```

```python
d = pd.read_excel('CAICavalanche.xls')
```

```python
sns.set()
fig, ax = plt.subplots()
ax.plot(d['MoSt'], d['SnowfallDiff'], 'bo', label = 'SnowfallDiff')
ax.plot(d['MoSt'], d['SnowdepthDiff'], 'mo', label = 'SnowdepthDiff')
ax.plot(d['MoSt'], d['lastyear']*200, 'co', label = 'lastyear')
ax.yaxis.set_visible(False)
ax.set_title('Snowfall and Avalanches: 2017', fontsize = 12)
ax.legend(loc='lower center', bbox_to_anchor=(0.5, -0.2), ncol=3)
plt.show()
```

Higher snowfall than average here is measured with two types of proxy variables: snowfall and snowdepth. A snowfall record here is set when the NOAA measure of snowfall on that day is larger than any snowfall measured $on$ $that$ $date$ in prior years. Snowfalldiff here is defined as the cumulative snowfall $above$ the prior record for all records set in that state in the given month. Snowdepth is calculated similarly. Each is calculated for the calendar month of the avalanche, for any NOAA observation in that state.

$lastyear$ is the number of avalanches in the same month in the prior avalanche year, for the same state.
